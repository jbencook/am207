# Estimating area

To estimate the area of the field, we need a function that tests whether an arbitrary point is inside the boundaries of the field. We could construct a polygon on the points and then count the number of times a line intersects a boundary to test whether a point is contained in the polygon. But since this field is defined by the points measured by a tractor, I simply test whether there are points further on all four corners of a given point to determine whether it is in the field.

```python
def valid_point(x):
    """                                                                         
    Test whether x and y are in within the bounds of the polygon                
    """
    tl = np.any(data[(data.lon < x[0]) & (data.lat < x[1])])
    tr = np.any(data[(data.lon > x[0]) & (data.lat < x[1])])
    bl = np.any(data[(data.lon < x[0]) & (data.lat > x[1])])
    br = np.any(data[(data.lon > x[0]) & (data.lat > x[1])])
    return np.all([tl, tr, bl, br])
```

Then, I use a Monte Carlo approach to esimate the area of the field. First I figured out how many miles a degree is both in longitude and latitude at this location in Nebraska. Then, I construct a rectangular box around the field and draw points uniformly inside of it. The proportion of points that fall in the field times the area of the rectangle is equal to the area of the field.

```python
#Get number of miles per degree for this longitude and latitude
x_range = (data.lon.max() - data.lon.min()) * 52
y_range = (data.lat.max() - data.lat.min()) * 69

sq_miles = y_range * x_range 

#Monte Carlo estimate
X2 = np.column_stack((np.random.uniform(data.lon.min(), data.lon.max(), 10000),
                 np.random.uniform(data.lat.min(), data.lat.max(), 10000)))

inside2 = np.zeros(10000)
for i in xrange(10000):
    inside2[i] = valid_point(X2[i,:])
    
#Number of acres
acres = sq_miles * inside2.mean() * 640
```

```python
inside2 = inside2.astype(bool)

fig = plt.figure(figsize=(5,5))
plt.plot(X2[inside2,0], X2[inside2,1], '.')
plt.plot(X2[~inside2,0], X2[~inside2,1], '.', c=".7")
plt.xlim([data.lon.min() - .001, data.lon.max() + .001])
plt.ylim([data.lat.min() - .001, data.lat.max() + .001])
plt.xticks([data.lon.min(), data.lon.max()], [])
plt.yticks([data.lat.min(), data.lat.max()], [])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("Monte Carlo estimate of area")
fig.savefig("mc_area.png")
```

# Pre-processing

Next, I define a grid of points for plotting and fit a k-nearest neighbors model to the points so that I can evaluate the yield function at any point in the field.

```python
# 100 x 100 grid
x_grid = np.linspace(data.lon.min(), data.lon.max(), 100)
y_grid = np.linspace(data.lat.min(), data.lat.max(), 100)

xx, yy = np.meshgrid(x_grid, y_grid)

xx = xx.flatten()
yy = yy.flatten()

# Drop points outside of observed polygon
mask = np.zeros(len(xx))
for i in xrange(len(xx)):
    tl = np.any(data[(data.lon < xx[i]) & (data.lat < yy[i])])
    tr = np.any(data[(data.lon > xx[i]) & (data.lat < yy[i])])
    bl = np.any(data[(data.lon < xx[i]) & (data.lat > yy[i])])
    br = np.any(data[(data.lon > xx[i]) & (data.lat > yy[i])])
    mask[i] = np.all([tl, tr, bl, br])

xx = xx[mask.astype(bool)]
yy = yy[mask.astype(bool)]

X = pd.DataFrame(np.column_stack((xx,yy)))
X.columns = ['lon', 'lat']
```

```python
knn = KNeighborsRegressor()
knn.fit(data[['lon', 'lat']], data['bu_ac'])
```

# Gaussian Process

Here, I sample $N_1 = 100$ points from the data and fit a Gaussian Process to them. I also define helper functions for predicting new points with the GP and returning the log of the prediction for the Metropolis-Hastings sampler.

```python
np.random.seed(1)
sample = data.loc[np.random.choice(np.arange(len(data)), 100, False)]

gp = GaussianProcess(theta0=2, nugget=1e-3)
gp.fit(sample[['lon', 'lat']], sample.bu_ac)

def predict(x,sample,K_inv,sigma2,phi,noise):
    if len(x.shape) == 1:
        K_star = cov(x.reshape(1,2), sample[['lon', 'lat']], sigma2, phi, noise)
    else:
        K_star = cov(x, sample[['lon', 'lat']], sigma2, phi, noise)
    return K_star.dot(K_inv).dot(sample['bu_ac'])

def log_f(x):
    return np.log(gp.predict(x) * valid_point(x))
```

```python
# Set up Metropolis Hastings 
def mh(N_mc, gamma, b, m):
    
    samples = np.empty((N_mc, 2))
    
    x1_prev, x2_prev = (data.lon.mean(), data.lat.mean())
    
    t = 0
    
    for i in xrange(b + m*N_mc):
        u = np.random.uniform()
        x1_star, x2_star = np.random.normal([x1_prev, x2_prev], gamma * np.ones(2))
        
        x_star = np.array([x1_star, x2_star])
        x_prev = np.array([x1_prev, x2_prev])
        p_star = log_f(x_star)
        p_prev = log_f(x_prev)
        
        if np.log(u) < p_star - p_prev:
            if i > b and i % m == 0:
                samples[t,:] = x1_star, x2_star
                t += 1
            x1_prev, x2_prev = x1_star, x2_star
        elif i > b and i % m == 0:
            samples[t,:] = x1_prev, x2_prev
            t += 1
            
    return samples[:-1,:]
```

```python
N_mc = 1000
gamma = .002
b = 100
m = 3
samples = mh(N_mc, gamma, b, m)
```
