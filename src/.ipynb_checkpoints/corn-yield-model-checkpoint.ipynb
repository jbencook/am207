# Introduction

Unprecedented amounts of data are available on modern farms. From yield monitors to weather sensors to infrared imaging, farmers are able to keep track of every detail on their farms. However, most farmers are not taking advantage of this data. Much of the data is never reviewed after being collected. The data that is reviewed remains inaccessible, trapped in complicated legacy software, agronomist reports, and countless pages of spreadsheets.

Agricultural production lags far behind other industries in using data to make decisions, but data-based decision-making will be essential in the future. Farming is a challenging business. Reduced availability of water and increasing input costs are squeezing farmers’ bottom lines during a time of extreme volatility in the commodities markets. New regulations on irrigation, pesticide use, and fertilizer application mean farmers must find a new way to boost yield. Soon, using data to optimize decision-making will no longer be a novel luxury – it will be essential for farmers to stay in business.

The purpose of this project is to develop and assess a procedure for predicting corn yield at a sub-field level before it has been harvested. Giving farmers insight about how much yield to expect in their fields will arm them to effectively compete against professional traders in the futures market. 

The approach uses the ideas of importance sampling to measure the high yield areas of a field and then models yield throughout the field accordingly. After $N$ training samples are drawn according to this importance sampling procedure, yield is modeled throughout the entire field with a Gaussian process.

In order to build a procedure that is robust and effective, we need to address the following questions:

1. How many points do we need to sample on the first pass to estimate the high yield areas?
2. On the first pass, should the points be sampled on a grid or randomly?
3. After the first pass, what is the best way to derive an importance sampling function?
4. What is the best method for drawing samples from the importance sampling function?
5. On the second pass, how many points do we need to sample to achieve an acceptably low level of variance?

Below, I sketch an outline of the procedure I will follow for this project including the data I am using, pre-processing and the method for estimating yield throughout the entire field. I end with a brief list of the next steps for my project.

# Data

The data come from a corn field in Butler County, Nebraska that was planted and harvested in 2009. Measurements are logged by a grain yield monitor which is connected to sensors on the arms of the tractor as it harvests the corn. The yield monitor records an estimate of yield in bushels per acre and geolocation approximately six times per second. This results in $N = 16,898$ points for a quarter mile$^2$, or 160-acre field.

```python
data = pd.read_csv("../data/wmk5_2009.csv")
print data.head()

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.scatter(data.lon, data.lat, marker='o', s=7, linewidths=0, c=data.bu_ac, cmap='YlGn')
cb = plt.colorbar()
plt.xlim([data.lon.min() - .001, data.lon.max() + .001])
plt.ylim([data.lat.min() - .001, data.lat.max() + .001])
plt.xticks([data.lon.min(), data.lon.max()], [data.lon.min(), data.lon.max()])
plt.yticks([data.lat.min(), data.lat.max()], [data.lat.min(), data.lat.max()])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("yield map")

plt.subplot(1, 2, 2)
plt.hist(data.bu_ac, bins=30, normed=True)
plt.xlabel("bushels per acre")
plt.title("distribution of yield")
plt.show()
```

One difficulty in analyzing these data is that the points are dense and irregularly spaced. At times during harvest, the tractor drove over the same spot more than once, so two points in almost the same location have very different estimates of yield. In order to compensate for this problem, a k-Nearest Neighbors regressor is fit to the data as a pre-processing step. This allows the recovery of a 100 $\times$ 100 point grid with a somewhat stronger signal. Points outside of the polygon of observed data are dropped.

```python
# 100 x 100 grid
x_grid = np.linspace(data.lon.min(), data.lon.max(), 100)
y_grid = np.linspace(data.lat.min(), data.lat.max(), 100)

xx, yy = np.meshgrid(x_grid, y_grid)

xx = xx.flatten()
yy = yy.flatten()

# Drop points outside of observed polygon
mask = np.zeros(len(xx))
for i in xrange(len(xx)):
    tl = np.any(data[(data.lon < xx[i]) & (data.lat < yy[i])])
    tr = np.any(data[(data.lon > xx[i]) & (data.lat < yy[i])])
    bl = np.any(data[(data.lon < xx[i]) & (data.lat > yy[i])])
    br = np.any(data[(data.lon > xx[i]) & (data.lat > yy[i])])
    mask[i] = np.all([tl, tr, bl, br])

xx = xx[mask.astype(bool)]
yy = yy[mask.astype(bool)]

# Fit kNN regressor to filter out bad data
knn = KNeighborsRegressor()
knn.fit(data[['lon', 'lat']], data.bu_ac)
zz = knn.predict(np.column_stack((xx,yy)))
```

For the most part, this pre-processing step does not seriously affect the yield map or the distribution of yield. The two differences are now we no longer have overlapping points in the yield map and we have fewer outliers in the distribution.

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.scatter(xx, yy, marker='o', s=8, linewidths=0, c=zz, cmap='YlGn')
plt.plot()
cb = plt.colorbar()
plt.xlim([data.lon.min() - .001, data.lon.max() + .001])
plt.ylim([data.lat.min() - .001, data.lat.max() + .001])
plt.xticks([data.lon.min(), data.lon.max()], [data.lon.min(), data.lon.max()])
plt.yticks([data.lat.min(), data.lat.max()], [data.lat.min(), data.lat.max()])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("pre-processed yield map")
plt.subplot(1, 2, 2)
plt.hist(zz, bins=30, normed=True)
plt.xlabel("bushels per acre")
plt.title("pre-processed distribution of yield")
plt.show()
```

# Method

The first step is to collect $N_1$ yield samples from the field. This corresponds to someone actually walking the field and collecting stalks at $N_1$ locations throughout the field. These locations could be chosen to form a grid or they could be randomly selected. Ideally, the procedure will be robust to the locations of this first pass since it is likely to be done manually.

It is important to note that I am currently making a strong simplifying assumption here: we can evaluate the actual yield function at these locations. That is, we can use the $N_1$ corn stalks to evaluate yield in bushels per acre at the sampled locations. In reality, inferring yield in bushels per acre based on volume of the grain from these $N_1$ stalks or based on height and circumference of the stalks will require a separate step.

```python
N_1 = 100

#Rescale data to lie in [0,1]
X = pd.DataFrame(rescale_data(np.column_stack((xx,yy)), min_val=0))
X.columns = ['zlon', 'zlat']
zz = rescale_data(zz, min_val=0)
X['zyield'] = zz
X = X.sort(columns=['zlon', 'zlat'])

#Randomly select N_1 points
indices = np.random.randint(0, len(X), N_1)
sample = X.loc[indices]

plt.figure(figsize=(5,5))
plt.scatter(X.zlon, X.zlat, marker='o', s=8, linewidths=0, c=X.zyield, cmap='YlGn')
cb = plt.colorbar()
plt.plot(sample.zlon, sample.zlat, 'o', c='black')
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("yield map with sample points overlaid")
plt.show()
```

The next step is to define an importance sampling function. We need something that is similar to the true yield function so that we concentrate more of our samples in high yield areas. One simple way to recover a function that is close to the true yield function is to fit a Gaussian process to the $N_1$ samples.

Here, I somewhat arbitrarily select hyperparameters, fit the Gaussian process to the $N_1$ samples and plot the importance sampling function.

```python
# First, a couple helper functions
def squared_exponential(d, sigma2, phi, noise):
    """
    Squared exponential kernel
    """
    if noise is not None:
        return sigma2 * np.exp( -np.square(d) / phi) + noise * np.eye(len(d))
    else:
        return sigma2 * np.exp( -np.square(d) / phi)

def kernel(*args, **kwargs):
    """
    Call squared_exponential kernel. 
    Takes either one design matrix or two in order to define K and K'
    """
    if len(args) == 1:
        X = args[0]
        d = spatial.distance_matrix(X, X)
    elif len(args) == 2:
        Xi, Xj = args
        d = spatial.distance_matrix(Xi, Xj)
    else:
        raise ValueError("incorrect number of arguments")
        
    sigma2 = kwargs['sigma2'] if 'sigma2' in kwargs else 1
    phi = kwargs['phi'] if 'phi' in kwargs else 1
    noise = kwargs['noise'] if 'noise' in kwargs else None
        
    return squared_exponential(d, sigma2, phi, noise)


# Next, I set the hyperparameters and compute the covariance matrices
phi = .1
sigma2 = 2
noise = 1e-3

K = kernel(sample[['zlon', 'zlat']],  sigma2=sigma2, phi=phi, noise=noise)
K_prime = kernel(X[['zlon', 'zlat']], sample[['zlon', 'zlat']],  sigma2=sigma2, phi=phi)

# And predict yield at the "unseen" locatons
y_hat = K_prime.dot(np.linalg.pinv(K)).dot(sample.zyield)

# Plot the importance sampling function
plt.figure(figsize=(5,5))
plt.scatter(X.zlon, X.zlat, marker='o', s=5, linewidths=0, c=y_hat, cmap='YlGn')
cb = plt.colorbar()
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("p(x, y)")
plt.show()
```

Next, we need to sample from this importance sampling function $p(x,y)$. I start with a simple Metropolis sampler with symmetric proposals. Unfortunately, this basic sampler is exhibiting random walk behavior. The Metropolis algorithm is unable to explore the entire space because of all the local maxima.

```python
#Helper functions
def valid_point(x):
    """
    Test whether x and y are in within the bounds of the polygon
    """
    tl = np.any(X[(X.zlon < x[0,0]) & (X.zlat < x[0,1])])
    tr = np.any(X[(X.zlon > x[0,0]) & (X.zlat < x[0,1])])
    bl = np.any(X[(X.zlon < x[0,0]) & (X.zlat > x[0,1])])
    br = np.any(X[(X.zlon > x[0,0]) & (X.zlat > x[0,1])])
    return np.all([tl, tr, bl, br])

def p(x):
    """
    Evaluate the importance sampling function
    """
    if valid_point(x):
        K_prime = kernel(x, sample[['zlon', 'zlat']], sigma2=sigma2, phi=phi)
        return K_prime.dot(np.linalg.pinv(K)).dot(sample.zyield)
    else:
        return np.zeros(1)

# Set up Metropolis Hastings 
N_mc = 1000
gamma = .02

samples = np.empty((N_mc, 2))
accepted = np.zeros(N_mc)

x1_prev, x2_prev = (.5, .5)

for i in xrange(N_mc):
    u = np.random.uniform()
    x1_star, x2_star = np.random.normal([x1_prev, x2_prev], gamma * np.ones(2))
    
    x_star = np.array([[x1_star, x2_star]])
    x_prev = np.array([[x1_prev, x2_prev]])
    p_star = np.log(p(x_star))
    p_prev = np.log(p(x_prev))
    
    if u < p_star / p_prev:
        samples[i,:] = x1_star, x2_star
        x1_prev, x2_prev = x1_star, x2_star
        accepted[i] = 1
    else:
        samples[i,:] = x1_prev, x2_prev
        
plt.figure(figsize=(5,5))

plt.scatter(X.zlon, X.zlat, marker='o', s=5, linewidths=0, c=y_hat, cmap='YlGn')
plt.plot(samples[:,0], samples[:,1], '.')
cb = plt.colorbar()
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("proposal function with M-H sampled points")

plt.show()
```

After sampling $N_{mc}$ points from the importance sampling function, I can "evaluate" the true yield function again by running the k-Nearest Neighbors regressor with the full dataset above to simulate resampling corn stalks from the actual field. Finally, I can estimate yield in the entire field $\int_{\Omega}f(x,y)dx dy \approx \sum_{i}f(x_i, y_i) / p(x_i, y_i)$ where the $x_i, y_i$ are sampled from $p(.)$.

```python
# Evaluate samples with importance sampling function
y_samples = []
for i in xrange(len(samples)):
    samp = p(np.array([samples[i,:]]))
    y_samples.append(samp[0])
    
p_samples = np.array(y_samples)

# Evaluate true yield function
knn.fit(X[['zlon', 'zlat']], X.zyield)
f_samples = knn.predict(samples)

# Approximate the integral
print np.sum(f_samples[p_samples != 0] / p_samples[p_samples != 0])
```
